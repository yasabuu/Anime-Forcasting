```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Untitled spreadsheet - Filtered_Animedataset_to_TV.csv')
df
```

```python
# Convert 'start_date' to datetime format if it's not already
df['Start_date'] = pd.to_datetime(df['Start_date'], format='%m/%d/%Y')

# Extract the year and add it as a new column 'year'
df['year'] = df['Start_date'].dt.year

df['month'] = df['Start_date'].dt.month

df['day'] = df['Start_date'].dt.day
```

```python
df
df['Genres'] = df['Genres'].str.split(',') # split by comma
df_exploded = df.explode('Genres')
df['Genres'] = df['Genres'].str.strip() # white spaces gone

df_exploded
```

```python
df_exploded.to_csv('/Filtered_exploded_genre_dates.csv', index=False)
```
