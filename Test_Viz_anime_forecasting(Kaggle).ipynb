```python
# group kaggle_df_exploded by genre
genre_id_count = kaggle_df_exploded.groupby('Genres')['anime_id'].nunique().reset_index()
genre_id_count
```

```python
genre_stats = kaggle_df_exploded.groupby('Genres').agg(
    anime_id_count=('anime_id', 'nunique'),
    score_avg = ('Score', 'mean'),
    popularity_sum=('Popularity', 'sum')
).reset_index()
genre_stats = genre_stats.sort_values(by='popularity_sum', ascending=False)
genre_stats
```

### **Machine Learning Models**

```python
df_genres_matrix = pd.get_dummies(df, columns=['Genres'], prefix='', prefix_sep='').groupby('anime_id').sum()
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Perform hierarchical clustering
Z = linkage(df_genres_matrix, method='ward')

# Plot a dendrogram to visualize clusters
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=df_genres_matrix.index)
plt.title("Dendrogram of Anime Genre Clusters")
plt.xlabel("Anime")
plt.ylabel("Distance")
plt.show()
```
